## Raw USAPL SBD Data

```python
raw_USAPL_SBD = open_powerlifting_full[
(open_powerlifting_full['Equipment'] == 'Raw') & 
(open_powerlifting_full['Federation'] == 'USAPL') &
(open_powerlifting_full['Event'] == 'SBD')
].reset_index(drop=True)

raw_USAPL_SBD = raw_USAPL_SBD.drop(columns=['Federation', 'ParentFederation', 'Tested', 'Event', 'Equipment', 'Wilks', 'Glossbrenner'])
```
